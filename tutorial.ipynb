# NRTDamage Tutorial

This notebook demonstrates how to use the 'NRTDamage' library.
It shows how to list materials, get their properties, and compute DPA.

```python
# Import the NRTDamage library
from nrtdamage.physics import compute_dpa
from nrtdamage.materials import get_material, list_materials

# List all available materials
print('Available materials:', list_materials())

# Get properties of iron
iron = get_material('iron')
print('\nIron properties:', iron)

# Compute DPA for iron after 1 year in a reactor
ONE_YEAR = 3.15e7  # seconds
dpa = compute_dpa(
    flux=1e14,
    cross_section=iron['cross_section'],
    damage_energy=1000.0,
    displacement_energy=iron['displacement_energy'],
    time=ONE_YEAR,
)
print(f'\nDPA after 1 year in iron: {dpa:.4e}')
```
